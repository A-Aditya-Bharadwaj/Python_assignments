```python
def Matrix_Multiplication(M1, M2):
    rows = len(M1)
    cols = len(M2[0])
    result = [[0]*cols for i in range(rows)]

    for i in range(len(M1)):
        for j in range(len(M2[0])):
            for k in range(len(M2)):
                result[i][j] = result[i][j] + (M1[i][k] * M2[k][j])

    return result
```

```python
X = [[12,7,3],
    [4 ,5,6],
    [7 ,8,9]]

Y = [[5,8,1,2],
    [6,7,3,0],
    [4,5,9,1]]

Matrix_Multiplication(X,Y)
```
