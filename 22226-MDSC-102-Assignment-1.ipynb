```python
#(a) f(x) = e^−x, x ≥ 0.
"""_Summary_
Finding E(x), E(x^2), Var(x) and E(X^n) for the function f(x) = e^−x, x ≥ 0.
"""
import sympy as sy
import math  

def f(x):
    return (math.e**(-x))

def exp_x():  
    x = sy.Symbol("x")
    return( sy.integrate((x*f(x)), (x, 0, sy.oo))) # integrating usnig sympy

def exp_x2():
    x = sy.Symbol("x")
    return( sy.integrate((x**2)*(f(x)), (x, 0, sy.oo)))

def varience():
    var = exp_x2() - (exp_x()**2)
    return var

def exp_xn():
    n = int(input("Give the value for n :"))
    x = sy.Symbol("x")
    return( sy.integrate((x**n)*(f(x)), (x, 0, sy.oo)))    

exp_x()
```
